### Folium 지도 시각화

```python
import folium
folium.__version__
```

```python
# Open street map에 광주광역시청을 표시
map = folium.Map(location=[35.159841, 126.851784], zoom_start=12) #위도, 경도 순 #zoom_start : 줌을 얼마에서 시작할건지
folium.Marker(
    location=[35.159841, 126.851784],
    popup=folium.Popup('광주광역시 서구 내방로 111', max_width=200), #popup : 마커를 클릭하면 뜨는 내용 #max_width : 최대너비설정
    tooltip='광주광역시청' #tootip : 마커 위에 커서를 대면 뜨는 내용
).add_to(map)
folium.Marker(
    location=[35.146107, 126.923314],
    popup=folium.Popup('광주광역시 동구 서남로 1', max_width=200), 
    tooltip='광주광역시 동구청' 
).add_to(map)
folium.Marker(
    location=[35.152089, 126.889858],
    popup=folium.Popup('광주광역시 서구 농성동 경열로 33', max_width=200), 
    tooltip='광주광역시 서구청' 
).add_to(map)
folium.Marker(
    location=[35.174100, 126.912157],
    popup=folium.Popup('광주광역시 북구 우치로 77', max_width=200), 
    tooltip='광주광역시 북구청' 
).add_to(map)
folium.Marker(
    location=[35.132900, 126.902573],
    popup=folium.Popup('광주광역시 남구 봉선로 1', max_width=200), 
    tooltip='광주광역시 남구청' 
).add_to(map)
folium.Marker(
    location=[35.139551, 126.793806],
    popup=folium.Popup('광주광역시 광산구 광산로29번길 15', max_width=200), 
    tooltip='광주광역시 광산구청' 
).add_to(map)
map
```

```python
map.save('광주시_주요관공서.html')
```

```python
# Stamen Toner map에 광주광역시청과 구청을 표시
from folium.features import CustomIcon, DivIcon

map = folium.Map(location=[35.159841, 126.851784], tiles="Stamen Toner", zoom_start=12) #Stamen Toner : 흑백배경으로 만들기
folium.Marker(
    location=[35.159841, 126.851784],
    popup=folium.Popup('광주광역시 서구 내방로 111', max_width=200), 
    tooltip='광주광역시청',
    icon=CustomIcon('광주광역시청.jpg', icon_size=(60,100))
).add_to(map)
folium.Marker(
    location=[35.146107, 126.923314],
    popup=folium.Popup('광주광역시 동구 서남로 1', max_width=200), 
    tooltip='광주광역시 동구청', 
    icon=DivIcon(icon_size=(100,20), icon_anchor=(0,0),
                 html='<div style="font-size: 14pt; background-color: powderblue"><bold>광주 동구청</bold></div>')
).add_to(map)
folium.Circle(
    radius=250, #반지름
    location=[35.152089, 126.889858],
    popup=folium.Popup('광주광역시 서구 농성동 경열로 33', max_width=200), 
    tooltip='광주광역시 서구청',
    color='magenta',
    fill=True 
).add_to(map)
folium.CircleMarker(
    radius=20,
    location=[35.174100, 126.912157],
    popup=folium.Popup('광주광역시 북구 우치로 77', max_width=200), 
    tooltip='광주광역시 북구청',
    color="#3186aa", #RGB, 16진수
    fill=True,
    fill_color="#3186dd"
).add_to(map)
folium.Marker(
    location=[35.132900, 126.902573],
    popup=folium.Popup('광주광역시 남구 봉선로 1', max_width=200), 
    tooltip='광주광역시 남구청',
    icon=folium.Icon('darkred', icon='star')
).add_to(map)
folium.Marker(
    location=[35.139551, 126.793806],
    popup=folium.Popup('광주광역시 광산구 광산로29번길 15', max_width=200), 
    tooltip='광주광역시 광산구청',
    icon=folium.Icon('blue', icon='plane')
).add_to(map)

title_html = '<h3 align="center">광주광역시 주요 행정기관</h3>' #h3 : 글씨체 중 하나 #align="center" : 가운데정렬
map.get_root().html.add_child(folium.Element(title_html)) #title 만들고 map이랑 연결
map
```
